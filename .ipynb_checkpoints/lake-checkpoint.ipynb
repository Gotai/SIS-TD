```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output

import gymnasium as gym
env = gym.make('FrozenLake-v1', desc=None, map_name="4x4", is_slippery=True , render_mode="rgb_array")
```

```python
""" 
Defaults https://gymnasium.farama.org/environments/toy_text/frozen_lake/

"4x4":[
        "SFFF",
        "FHFH",
        "FFFH",
        "HFFG"
        ]

    "8x8": [
        "SFFFFFFF",
        "FFFFFFFF",
        "FFFHFFFF",
        "FFFFFHFF",
        "FFFHFFFF",
        "FHHFFFHF",
        "FHFFHFHF",
        "FFFHFFFG",
    ]

"""
```

```python
# States will be tuples as we are lookin at NxN Matrices for our environment

class LakeRunner:
    def __init__(self, learning_rate, epsilon, decay, final_epsilon, gamma = .9):
        # create Q-Table, we need a (B, T)
        self.q_table = np.zeros([env.observation_space.n, env.action_space.n])
        self.lr = learning_rate
        self.gamma = gamma
        self.epsilon = epsilon
        self.decay = decay
        self.final_epsilon = final_epsilon
        self.delta = []

    def greedy_pi(self, state):
        if np.random.random() < self.epsilon:
            return env.action_space.sample()
        else:
            return int(np.argmax(self.q_table[state]))
    
    # note reward is actually reward_t+1
    def update(self, state, action, reward, next_state):
        # Spliting this into error and target, so we can track the target

        error =  reward + self.gamma * np.max(self.q_table[next_state]) - self.q_table[state][action]
        self.q_table[state][action] = self.q_table[state][action] + self.lr * error
        self.delta.append(error)

    def decay_epsilon(self):
        self.epsilon = max(self.final_epsilon, self.epsilon - self.decay)
```

```python
# hyperparameters
learning_rate = 0.01
max_iter = 1000
epsilon = 1.0
decay = epsilon / (max_iter/ 2)  # reduce the exploration over time
final_epsilon = 0.1

agent = LakeRunner(learning_rate, epsilon, decay, final_epsilon)
```

```python
#env.reset()
#env = gym.wrappers.RecordEpisodeStatistics(env, deque_size=max_iter)
for episode in range(max_iter):
    state, info = env.reset()
    done = False
    clear_output()

    # play one episode
    while not done:
        action = agent.greedy_pi(state)
        next_state, reward, terminated, truncated, info = env.step(action)
        
        # update the agent
        agent.update(state, action, reward, next_state)
        frame = env.render()
        plt.imshow(frame)
        plt.show()
        # update if the environment is done and the current state
        done = terminated or truncated
        state = next_state

    agent.decay_epsilon()
```
